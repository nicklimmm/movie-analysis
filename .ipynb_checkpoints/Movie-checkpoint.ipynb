```python
def get_movie(movie_id):
    details = requests.get(f'https://api.themoviedb.org/3/movie/{movie_id}?api_key={API_KEY}&language=en-US').json()
    casts = requests.get(f'https://api.themoviedb.org/3/movie/{movie_id}/credits?api_key={API_KEY}&language=en-US').json()
    details.update(casts)
    
    return details
```

```python
def get_all_movie():
    movie_id = 1
    while True:
        details = requests.get(f'https://api.themoviedb.org/3/movie/{movie_id}?api_key={API_KEY}&language=en-US').status_code
        if details != 200:
            print(movie_id)
            break
        get_movie(movie_id)
```
