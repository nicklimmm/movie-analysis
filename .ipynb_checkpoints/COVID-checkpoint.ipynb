```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import requests
import seaborn as sns
```

```python
covid_df = pd.read_csv('datasets/daily.csv')
covid_df.head()
```

```python
covid_df.tail()
```

```python
covid_df.info()
```

```python
covid_df = covid_df.drop(['recovered', 'total', 'posNeg', 'hash', 'dateChecked', 'lastModified'], axis=1)
covid_df = covid_df.fillna(0)

covid_df['date'] = pd.to_datetime(covid_df['date'], format='%Y%m%d')
covid_df = covid_df.set_index('date')

covid_df = covid_df.astype('int64')

covid_df
```

```python
covid_df.info()
```
